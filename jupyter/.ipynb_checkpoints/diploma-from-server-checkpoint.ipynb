```python
import pandas as pd
from gensim.models import KeyedVectors
from gensim.corpora import Dictionary

def normalize(phrase):
    if '(' in phrase:
        phrase = phrase[:phrase.index('(')]
    return '_'.join(phrase.lower().split())
```

```python
folder = 'virtual_reality'
# folder = 'severstal'

if folder == 'virtual_reality':
    params = {
        'index_col': 'Unnamed: 0'
    }
else:
    params = {}

# corpus_occurence = pd.read_csv('./data/%s/corpus_occurrence_by_year.csv' % folder)
cooccurrence = pd.read_csv('./data/%s/cooccurrence_by_year.csv' % folder, **params)
occurrence = pd.read_csv('./data/%s/occurrence_by_year.csv' % folder, **params)
specificity = pd.read_csv('./data/%s/specificity_statistics.csv' % folder, **params)
```

```python
%%time
word_vectors = KeyedVectors.load_word2vec_format('./data/ngram.model.bin', binary=True)
```

```python
vocab = set(word_vectors.vocab)
len(vocab)
```

```python
set1 = set(cooccurrence.ngram1) | set(cooccurrence.ngram2)
set2 = set(specificity.ngram)
set3 = set(occurrence.ngram)
ngrams = set1 & set2 & set3
len(set1), len(set2), len(set3), len(ngrams)
```

```python
data = {}
data_by_ngram = {}
not_matched = 0
for ngram in ngrams:
    token = normalize(ngram)
    if token not in vocab:
        not_matched += 1
        continue

    entry = {
        'ngram': ngram,
        'token': token,
        'vector': word_vectors.get_vector(token).tolist(),
        'cooc_by_year': {},
        'cooc': {},
        'oc_by_year': {},
        'oc': {},
    }
    if token in data:
        print(ngram, '|', data[token]['ngram'], '|', token)
    data_by_ngram[ngram] = entry
    data[token] = entry

print()
print('%d not matched' % not_matched)
print('%d matched' % len(data))
```

```python
for no, (ngram1, ngram2, pubyear, count) in cooccurrence.iterrows():
    if ngram1 not in data_by_ngram or ngram2 not in data_by_ngram:
        continue

    entry1 = data_by_ngram[ngram1]
    entry2 = data_by_ngram[ngram2]
    
    token1 = entry1['token']
    token2 = entry2['token']

    cooc_by_year1 = entry1['cooc_by_year']
    cooc_by_year2 = entry2['cooc_by_year']

    if pubyear not in cooc_by_year1:
        cooc_by_year1[pubyear] = {}
    if pubyear not in cooc_by_year2:
        cooc_by_year2[pubyear] = {}
    
    cooc_by_year1[pubyear][token2] = count
    cooc_by_year2[pubyear][token1] = count
    
for entry in data.values():
    entry['cooc'] = {}
    for year in entry['cooc_by_year']:
        for token2, count in entry['cooc_by_year'][year].items():
            if token2 not in entry['cooc']:
                entry['cooc'][token2] = 0
            entry['cooc'][token2] += count
```

```python
for no, (ngram, pubyear, count, rel_freq) in occurrence.iterrows():
    if ngram not in data_by_ngram:
        continue
    
    entry = data_by_ngram[ngram]

    entry['oc_by_year'][pubyear] = {
        'count': count,
        'rel_freq': rel_freq,
    }
    
for entry in data.values():
    entry['oc'] = 0
    for year in entry['oc_by_year']:
        e = entry['oc_by_year'][year]
        entry['oc'] += e['count']
```

```python
for no, (id_, ngram, ngramids, corpus_sum, ngram_length, specificity, in_degree, out_degree, cluster, n_types) in specificity.iterrows():
    if ngram not in data_by_ngram:
        continue
    
    entry = data_by_ngram[ngram]
    entry['in_degree'] = in_degree
    entry['out_degree'] = out_degree
    entry['cluster'] = cluster
    entry['n_types'] = n_types
    entry['specificity'] = specificity
```

```python
import cairo

FONTSIZE = 1024

def draw_label(ctx, x, y, text, extents, size, stroke=False, rotate=0):
    (bx, by, w, h, ax, ay) = extents
    ctx.save()
    ctx.translate(x, y)
    s = size / FONTSIZE
    if rotate:
        ctx.rotate(rotate)
    ctx.scale(s, s)
    ctx.move_to(- w / 2 - bx, - h / 2 - by)
    if stroke:
        ctx.text_path(text)
        ctx.stroke()
    else:
        ctx.show_text(text)
    ctx.restore()
    
def label_bbox(x, y, extents, size):
    (bx, by, w, h, ax, ay) = extents
    s = size / FONTSIZE
    w *= s
    h *= s
    bx *= s
    by *= s
    return (x, y, w, h)

surface = cairo.ImageSurface(cairo.FORMAT_ARGB32, FONTSIZE, FONTSIZE)
ctx = cairo.Context(surface)
ctx.select_font_face("", cairo.FONT_SLANT_NORMAL, cairo.FONT_WEIGHT_NORMAL)
ctx.set_font_size(FONTSIZE)
for entry in data.values():
    entry['text'] = entry['ngram']
    entry['extents'] = ctx.text_extents(entry['text'])
for text in sphere_texts:
    sphere_extents.append(ctx.text_extents(text))
```

```python
import math

def draw_tsne2d(vec2d, fname):
    xmax, ymax = vec2d.max(axis=0)
    xmin, ymin = vec2d.min(axis=0)
    
    SIDE = 5000
    
    def proj(x, y):
        return (
            (x - xmin) / (xmax - xmin) * SIDE,
            (y - ymin) / (ymax - ymin) * SIDE,
        )
    
    surface = cairo.ImageSurface(cairo.FORMAT_ARGB32, SIDE, SIDE)
    ctx = cairo.Context(surface)
    ctx.select_font_face("", cairo.FONT_SLANT_NORMAL, cairo.FONT_WEIGHT_NORMAL)
    ctx.set_font_size(FONTSIZE)
    
    ctx.set_source_rgb(1.0, 1.0, 1.0)
    ctx.paint()

    ctx.set_source_rgba(0, 0, 0, 0.05)

    for i, token in enumerate(token_order):
        entry = data[token]
        vec0 = vec2d[i, :]
        x0, y0 = proj(vec0[0], vec0[1])
        for token2, count in entry['cooc'].items():
            if token >= token2:
                continue
            ctx.set_line_width(math.sqrt(count))
            j = token2pos[token2]
            vec1 = vec2d[j, :]
            x1, y1 = proj(vec1[0], vec1[1])
            ctx.move_to(x0, y0)
            ctx.line_to(x1, y1)
            ctx.stroke()
    
    for i, token in enumerate(token_order):
        entry = data[token]
        vec = vec2d[i, :]
        norm = np.linalg.norm(data[token]['vector'])
        x, y = proj(vec[0], vec[1])
        extents = entry['extents']
        size = 2 * norm # 10 * math.log(1 + entry['oc'])
        cx, cy, w, h = label_bbox(x, y, extents, size)
        ctx.set_source_rgb(1.0, 1.0, 1.0)
        ctx.set_line_width(FONTSIZE * 0.1)
        draw_label(ctx, x, y, entry['text'], extents, size, stroke=True)
        ctx.set_source_rgb(1.0, 0.0, 0.0)
        draw_label(ctx, x, y, entry['text'], extents, size)

    surface.write_to_png('%s.png' % fname)

draw_tsne2d(vec2d[30], 'tsne2d_30')
```

```python
import math
import matplotlib as mpl

cmap = mpl.cm.get_cmap('gist_ncar')

def draw_voronoi(vec2d, fname):
    xmax, ymax = vec2d.max(axis=0)
    xmin, ymin = vec2d.min(axis=0)
    
    SIDE = 5000
    
    def proj(x, y):
        return (
            (x - xmin) / (xmax - xmin) * SIDE,
            (y - ymin) / (ymax - ymin) * SIDE,
        )
    
    surface = cairo.ImageSurface(cairo.FORMAT_ARGB32, SIDE, SIDE)
    ctx = cairo.Context(surface)
    ctx.select_font_face("", cairo.FONT_SLANT_NORMAL, cairo.FONT_WEIGHT_NORMAL)
    ctx.set_font_size(FONTSIZE)
    
    ctx.set_source_rgb(1.0, 1.0, 1.0)
    ctx.paint()

    ctx.set_source_rgba(0, 0, 0, 1.0)

    vor = Voronoi(vec2d)
    region2point = {region: point for point, region in enumerate(vor.point_region)}
    
    for region_i, region in enumerate(vor.regions):
        if len(region) == 0:
            continue
        for vert in region:
            if vert == -1:
                break
        if vert == -1:
            continue
        for vert in region:
            x, y = proj(*vor.vertices[vert, :])
            ctx.line_to(x, y)
        point = region2point[region_i]
        val = np.searchsorted(vec1dsorted, vec1d[30][point][0]) / len(vec1dsorted)
        r, g, b, a = cmap(val)
        ctx.set_source_rgba(r, g, b, 0.3)
        ctx.fill()
    
    for (v1, v2), (p1, p2) in zip(vor.ridge_vertices, vor.ridge_points):
        if v1 == -1 or v2 == -1:
            continue
        x1, y1 = proj(*vor.vertices[v1, :])
        x2, y2 = proj(*vor.vertices[v2, :])
        vec1 = vec2d[p1, :]
        vec2 = vec2d[p2, :]
        norm = scipy.spatial.distance.cosine(vec1, vec2)
        weight = max(np.searchsorted(norms, norm) / len(norms) - 0.5, 0) * 10
        if weight == 0:
            continue
        ctx.set_source_rgba(0, 0, 0, 1)
        ctx.set_line_width(weight)
        ctx.move_to(x1, y1)
        ctx.line_to(x2, y2)
        ctx.stroke()
    
    for i, token in enumerate(token_order):
        entry = data[token]
        vec = vec2d[i, :]
        norm = np.linalg.norm(data[token]['vector'])
        x, y = proj(vec[0], vec[1])
        extents = entry['extents']
        size = 10 * math.log(1 + entry['oc']) # 2 * norm # 10 * math.log(1 + entry['oc'])
        cx, cy, w, h = label_bbox(x, y, extents, size)
        ctx.set_source_rgb(1.0, 1.0, 1.0)
        ctx.set_line_width(FONTSIZE * 0.1)
        draw_label(ctx, x, y, entry['text'], extents, size, stroke=True)
        ctx.set_source_rgb(0.0, 0.0, 0.0)
        draw_label(ctx, x, y, entry['text'], extents, size)

    surface.write_to_png('%s.png' % fname)

draw_voronoi(vec2d[30], 'voronoi')
```

```python
import math
import matplotlib as mpl

cmap = mpl.cm.get_cmap('gist_ncar')

def draw_freq_vs_norm(vec2d, fname):
    WIDTH = 10000
    HEIGHT = 5000
    
    norms = np.array([np.linalg.norm(entry['vector']) for entry in data.values()])
    norms.sort()
    min_norm, max_norm = norms.min(), norms.max()
    
    freqs = np.array([np.linalg.norm(entry['oc']) for entry in data.values()])
    freqs.sort()
    min_freq, max_freq = math.log(freqs.min()), math.log(freqs.max())
    
    def proj(freq, norm):
        return (
            (0.05 + 0.9 * (freq - min_freq) / (max_freq - min_freq)) * WIDTH,
            (1 - 0.05 - 0.9 * (norm - min_norm) / (max_norm - min_norm)) * HEIGHT,
        )
    
    surface = cairo.ImageSurface(cairo.FORMAT_ARGB32, WIDTH, HEIGHT)
    ctx = cairo.Context(surface)
    ctx.select_font_face("", cairo.FONT_SLANT_NORMAL, cairo.FONT_WEIGHT_NORMAL)
    ctx.set_font_size(FONTSIZE)
    
    ctx.set_source_rgb(1.0, 1.0, 1.0)
    ctx.paint()

    ctx.set_source_rgba(0, 0, 0, 1.0)

    for i, token in enumerate(token_order):
        entry = data[token]
        vec = vec2d[i, :]
        freq = math.log(entry['oc'])
        norm = np.linalg.norm(entry['vector'])
        x, y = proj(freq, norm)
        extents = entry['extents']
        size = 10
        cx, cy, w, h = label_bbox(x, y, extents, size)
        ctx.set_source_rgb(1.0, 1.0, 1.0)
        ctx.set_line_width(FONTSIZE * 0.1)
        draw_label(ctx, x, y, entry['text'], extents, size, stroke=True)
        ctx.set_source_rgb(1.0, 0.0, 0.0)
        draw_label(ctx, x, y, entry['text'], extents, size)

    surface.write_to_png('%s.png' % fname)

draw_freq_vs_norm(vec2d[30], 'freq_vs_norm')
```

```python
pt2px = ([x, y, z]) => {
const t = this.state.target;
const dx = x - t[0];
const dy = y - t[1];
const dz = z - t[2];

const [alpha, beta, dist, fov] = this.state.viewport;
const dx1 = dz * Math.sin(-alpha) + dx * Math.cos(-alpha);
const dy1 = dy;
const dz1 = dz * Math.cos(-alpha) - dx * Math.sin(-alpha);

const dx2 = dx1;
const dy2 = dz1 * Math.sin(-beta) + dy1 * Math.cos(-beta);
const dz2 = dz1 * Math.cos(-beta) - dy1 * Math.sin(-beta);

const dx3 = dx2;
const dy3 = dy2;
const dz3 = dist - dz2;

const h = Math.tan(fov / 2) * dist;
const c = dist / dz3 / h * this.props.height / 2;
const u = dx3 * c;
const v = dy3 * c;

return [u + this.props.width / 2, -v + this.props.height / 2, 1 / dz3];
};
```

```python
def gen_proj3d(target, viewport, width, height):
    def proj(pt):
        dx = pt[0] - target[0]
        dy = pt[1] - target[1]
        dz = pt[2] - target[2]
        alpha, beta, dist, fov = viewport
        dx1 = dz * math.sin(-alpha) + dx * math.cos(-alpha)
        dy1 = dy
        dz1 = dz * math.cos(-alpha) - dx * math.sin(-alpha)
        
        dx2 = dx1
        dy2 = dz1 * math.sin(-beta) + dy1 * math.cos(-beta)
        dz2 = dz1 * math.cos(-beta) - dy1 * math.sin(-beta)
        
        dx3 = dx2
        dy3 = dy2
        dz3 = dist - dz2
        
        h = math.tan(fov / 2) * dist
        c = dist / dz3 / h * height / 2
        u = dx3 * c
        v = dy3 * c
        
        return (
            u + width / 2,
            -v + height / 2,
            1 / dz3
        )
    return proj
```

```python
def mix(a, b, t):
    return a * (1 - t) + b * t
```

```python
def draw_tsne3d_to_tsne2d(vec3d, vec2d, t, fname, target, viewport):
    
    proj = gen_proj3d(target, viewport, WIDTH, HEIGHT)
    
    surface = cairo.ImageSurface(cairo.FORMAT_ARGB32, WIDTH, HEIGHT)
    ctx = cairo.Context(surface)
    ctx.select_font_face("", cairo.FONT_SLANT_NORMAL, cairo.FONT_WEIGHT_NORMAL)
    ctx.set_font_size(FONTSIZE)
    
    ctx.set_source_rgb(1.0, 1.0, 1.0)
    ctx.paint()
    
    def get_vec(i):
        x1, y1, z1 = vec3d[i, :]
        x2, y2 = vec2d[i, :]
        return (
            mix(x1, y2, t),
            mix(y1, -x2, t),
            mix(z1, 0, t),
        )
        
    projections = []
    for i, token in enumerate(token_order):
        vec = get_vec(i)
        x, y, s = proj(vec)
        if s < 0:
            continue
        projections.append((s, (x, y, s), i, token))
    projections.sort()
    
    ctx.set_source_rgba(0, 0, 0, 0.05)
    
    for _, (x, y, s), i, token in projections:
        entry = data[token]
        extents = entry['extents']
        size = 300 * s
        cx, cy, w, h = label_bbox(x, y, extents, size)
        ctx.set_source_rgb(1.0, 1.0, 1.0)
        ctx.set_line_width(FONTSIZE * 0.2)
        draw_label(ctx, x, y, entry['text'], extents, size, stroke=True)

        val = np.searchsorted(vec1dsorted, vec1d[30][i][0]) / len(vec1dsorted)
        r, g, b, a = cmap(val)
        ctx.set_source_rgb(r, g, b)
        draw_label(ctx, x, y, entry['text'], extents, size)

    surface.write_to_png('%s.png' % fname)
```

```python
def get_edge_pt(source, target, height, t):
    return (
        mix(source[0], target[0], t),
        t * (1 - t) * 4 * height,
        mix(source[2], target[2], t),
    )
```

```python
def draw_graph_over_tsne2d(vec2d, fname, target, viewport):
    
    proj = gen_proj3d(target, viewport, WIDTH, HEIGHT)
    
    surface = cairo.ImageSurface(cairo.FORMAT_ARGB32, WIDTH, HEIGHT)
    ctx = cairo.Context(surface)
    ctx.select_font_face("", cairo.FONT_SLANT_NORMAL, cairo.FONT_WEIGHT_NORMAL)
    ctx.set_font_size(FONTSIZE)
    
    ctx.set_source_rgb(1.0, 1.0, 1.0)
    ctx.paint()
        
    projections = []

    for i, token in enumerate(token_order):
        x, y = vec2d[i, :]
        pt1 = x, 0, y

        x1, y1, s1 = proj((x, 0, y))
        if s1 < 0:
            continue
        projections.append((s1, 'label', (x1, y1, i, token)))
        
        entry = data[token]
        for token1, count in entry['cooc'].items():
            x, y = vec2d[token2pos[token1], :]
            pt2 = x, 0, y
            dist = scipy.spatial.distance.euclidean(pt1, pt2)
            if count == 1:
                continue

            n = math.ceil(dist)
            for t in range(n):
                seg_pt_1 = proj(get_edge_pt(pt1, pt2, dist / 6, t / n))
                seg_pt_2 = proj(get_edge_pt(pt1, pt2, dist / 6, (t + 1) / n))
                sm = 2 / (1 / seg_pt_1[0] + 1 / seg_pt_2[0])
                projections.append((sm, 'line', seg_pt_1 + seg_pt_2 + (math.log(count),)))
    
    projections.sort()
    
    ctx.set_source_rgba(0, 0, 0, 0.05)
    
    for s, kind, params in projections:
        if kind == 'label':
            x, y, i, token = params
            entry = data[token]
            extents = entry['extents']
            size = 300 * s
            cx, cy, w, h = label_bbox(x, y, extents, size)
            ctx.set_source_rgb(1.0, 1.0, 1.0)
            ctx.set_line_width(FONTSIZE * 0.2)
            draw_label(ctx, x, y, entry['text'], extents, size, stroke=True)

            val = np.searchsorted(vec1dsorted, vec1d[30][i][0]) / len(vec1dsorted)
            r, g, b, a = cmap(val)
            ctx.set_source_rgb(r, g, b)
            draw_label(ctx, x, y, entry['text'], extents, size)

        elif kind == 'line':
            x1, y1, s1, x2, y2, s2, width = params
            ctx.set_source_rgba(0, 0, 1.0, 0.2)
            ctx.set_line_width(width * s1 * 10)
            ctx.move_to(x1, y1)
            ctx.line_to(x2, y2)
            ctx.stroke()

    surface.write_to_png('%s.png' % fname)
```

```python
def draw_tsne2d_to_circular(vec2d, t, fname, target, viewport):
    
    proj = gen_proj3d(target, viewport, WIDTH, HEIGHT)
    
    surface = cairo.ImageSurface(cairo.FORMAT_ARGB32, WIDTH, HEIGHT)
    ctx = cairo.Context(surface)
    ctx.select_font_face("", cairo.FONT_SLANT_NORMAL, cairo.FONT_WEIGHT_NORMAL)
    ctx.set_font_size(FONTSIZE)
    
    ctx.set_source_rgb(1.0, 1.0, 1.0)
    ctx.paint()
    
    def get_vertex_proj(i):
        x, y = vec2d[i, :]
        pt1 = x, y, 0
        
        u = np.searchsorted(vec1dsorted, vec1d[30][i][0])
        u /= len(vec1dsorted)
        u -= 0.5
        u *= 2 * math.pi
        r = 50
        pt2 = math.cos(u) * r, math.sin(u) * r, 0
        
        return proj((
            mix(pt1[0], pt2[0], t),
            mix(pt1[1], pt2[1], t),
            mix(pt1[2], pt2[2], t),
        )), u
        
        
    projections = []

    for i, token in enumerate(token_order):
        (x1, y1, s1), u = get_vertex_proj(i)
        if s1 < 0:
            continue
        projections.append((s1, 'label', (x1, y1, i, token, mix(0, -u, t))))
        entry = data[token]
        for token1, count in entry['cooc'].items():
            (x2, y2, s2), u2 = get_vertex_proj(token2pos[token1])
            if s2 < 0:
                continue
            if count == 1:
                continue
            projections.append((0, 'line', (x1, y1, s1, x2, y2, s2, math.log(count))))
            continue
            
            n = math.ceil(dist)
            for t in range(n):
                seg_pt_1 = proj(get_edge_pt(pt1, pt2, dist / 6, t / n))
                seg_pt_2 = proj(get_edge_pt(pt1, pt2, dist / 6, (t + 1) / n))
                sm = 2 / (1 / seg_pt_1[0] + 1 / seg_pt_2[0])
                projections.append((sm, 'line', seg_pt_1 + seg_pt_2 + (math.log(count),)))
    
    projections.sort()
    
    ctx.set_source_rgba(0, 0, 0, 0.05)
    
    for s, kind, params in projections:
        if kind == 'label':
            x, y, i, token, rotate = params
            entry = data[token]
            extents = entry['extents']
            size = 300 * s
            cx, cy, w, h = label_bbox(x, y, extents, size)
            ctx.set_source_rgb(1.0, 1.0, 1.0)
            ctx.set_line_width(FONTSIZE * 0.2)
            draw_label(ctx, x, y, entry['text'], extents, size, stroke=True, rotate=rotate)

            val = np.searchsorted(vec1dsorted, vec1d[30][i][0]) / len(vec1dsorted)
            r, g, b, a = cmap(val)
            ctx.set_source_rgb(r, g, b)
            draw_label(ctx, x, y, entry['text'], extents, size, rotate=rotate)

        elif kind == 'line':
            x1, y1, s1, x2, y2, s2, width = params
            ctx.set_source_rgba(0, 0, 1.0, 0.2)
            ctx.set_line_width(width * s1 * 10)
            ctx.move_to(x1, y1)
            ctx.line_to(x2, y2)
            ctx.stroke()

    surface.write_to_png('%s.png' % fname)
```

```python
def draw_tsne3d_sphere(vec3d, a1, a2, fname, target, viewport):
    
    proj = gen_proj3d(target, viewport, WIDTH, HEIGHT)
    
    surface = cairo.ImageSurface(cairo.FORMAT_ARGB32, WIDTH, HEIGHT)
    ctx = cairo.Context(surface)
    ctx.select_font_face("", cairo.FONT_SLANT_NORMAL, cairo.FONT_WEIGHT_NORMAL)
    ctx.set_font_size(FONTSIZE)
    
    ctx.set_source_rgb(1.0, 1.0, 1.0)
    ctx.paint()
    
    projections = []
    center = vec3d[token2pos[sphere_token], :]

    for i, token in enumerate(token_order):
        vec = vec3d[i, :]
        x, y, s = proj(vec)
        if s < 0:
            continue
        projections.append((s, (x, y, s), i, token, False))
        
    for i, ((x, y, z), token) in enumerate(zip(sphere_tsne2d, sphere_words)):
        sc = 0.01
        x, y, s = proj((center[0] + x * sc, center[1] + y * sc, center[2] + z * sc))
        if s < 0:
            continue
        projections.append((s, (x, y, s * 0.1), i, token, True))

    projections.sort()
    
    ctx.set_source_rgba(0, 0, 0, 0.05)
    
    for _, (x, y, s), i, token, in_sphere in projections:
        if in_sphere:
            extents = sphere_extents[i]
            text = sphere_texts[i]
        else:
            entry = data[token]
            vec = vec3d[i, :]
            norm = np.linalg.norm(data[token]['vector'])
            extents = entry['extents']
            text = entry['text']
        size = 300 * s
        cx, cy, w, h = label_bbox(x, y, extents, size)
        ctx.set_source_rgba(1.0, 1.0, 1.0, a2 if in_sphere else a1)
        ctx.set_line_width(FONTSIZE * 0.2)
        draw_label(ctx, x, y, text, extents, size, stroke=True)

        if in_sphere:
            ctx.set_source_rgba(0, 0, 0, a2)            
        else:
            val = np.searchsorted(vec1dsorted, vec1d[30][i][0]) / len(vec1dsorted)
            r, g, b, a = cmap(val)
            ctx.set_source_rgba(r, g, b, a1)
        draw_label(ctx, x, y, text, extents, size)

    surface.write_to_png('%s.png' % fname)
```

```python
def draw_tsne3d_kmeans(vec3d, t, ealpha, fname, target, viewport):
    
    proj = gen_proj3d(target, viewport, WIDTH, HEIGHT)
    
    surface = cairo.ImageSurface(cairo.FORMAT_ARGB32, WIDTH, HEIGHT)
    ctx = cairo.Context(surface)
    ctx.select_font_face("", cairo.FONT_SLANT_NORMAL, cairo.FONT_WEIGHT_NORMAL)
    ctx.set_font_size(FONTSIZE)
    
    ctx.set_source_rgb(1.0, 1.0, 1.0)
    ctx.paint()
    
    projections = []
    
    def get_point_proj(i):
        vec = vec3d[i, :]
        cluster = kmeans_clusters[i]
        cluster_center = kmeans.cluster_centers_[cluster]
        return proj(vec + cluster_center * t)

    for i, token in enumerate(token_order):
        x, y, s = get_point_proj(i)
        if s < 0:
            continue
        projections.append((s, 'label', (x, y, s, i, token, kmeans_clusters[i])))
        entry = data[token]
        for token2, count in entry['cooc'].items():
            x2, y2, s2 = get_point_proj(token2pos[token2])
            sm = 2 / (1 / s + 1 / s2)
            projections.append((sm, 'line', (x, y, s, x2, y2, s2, math.sqrt(count))))

    projections.sort()
    
    ctx.set_source_rgba(0, 0, 0, 0.05)
    
    for sm, kind, params in projections:
        if kind == 'label':
            x, y, s, i, token, cluster = params
            entry = data[token]
            vec = vec3d[i, :]
            norm = np.linalg.norm(data[token]['vector'])
            extents = entry['extents']
            text = entry['text']
            size = 300 * s
            cx, cy, w, h = label_bbox(x, y, extents, size)
            ctx.set_source_rgba(1.0, 1.0, 1.0, 1.0)
            ctx.set_line_width(FONTSIZE * 0.2)
            draw_label(ctx, x, y, text, extents, size, stroke=True)

            val = cluster / len(kmeans.cluster_centers_)
            r, g, b, a = cmap(val)
            ctx.set_source_rgba(r, g, b, 1.0)
            draw_label(ctx, x, y, text, extents, size)
        
        elif kind == 'line':
            x1, y1, s1, x2, y2, s2, width = params
            ctx.set_source_rgba(0, 0, 1.0, ealpha)
            ctx.set_line_width(width * sm * 5)
            ctx.move_to(x1, y1)
            ctx.line_to(x2, y2)
            ctx.stroke()

    surface.write_to_png('%s.png' % fname)
```

```python
folder = '8'
fname = 'frames/%s' % folder
if not os.path.exists(fname):
    os.makedirs(fname)

frames1 = 100
for frame in tqdm(range(frames1)):
    t = frame / (frames1 - 1)
    target = (0, 0, 0)
    dist = mix(70, 150, smooth(t))
    alpha = 0.5 * t
    beta = 0
    draw_tsne3d_kmeans(vec3d, mix(0, 1.5, smooth(t)), mix(0, 0.8, t), 'frames/%s/%06d' % (folder, frame), target, (alpha, beta, dist, math.pi / 2))

frames2 = 100
for frame in tqdm(range(frames2)):
    t = frame / (frames2 - 1)
    target = (0, 0, 0)
    dist = 150
    alpha = 0.5 + 0.5 * t
    beta = 0
    draw_tsne3d_kmeans(vec3d, 1.5, 0.8, 'frames/%s/%06d' % (folder, frames1 + frame), target, (alpha, beta, dist, math.pi / 2))
```
